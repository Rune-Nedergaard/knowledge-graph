After playing with NER method below, it became clear that some "navn" have written "udpeget af" in them, which registers as institution by ner. Such as: (Pia Gjellerup (udpeget af S))

To avoid this, we first find all that have "udpeget af" in them, and then remove everything after the first "(" along with all parantheses. This is done by the function "remove_udpeget_af"

```python

import regex as re
import nltk as nltk

#After playing with NER method below, it became clear that some "navn" have written "udpeget af" in them, which registers as institution by ner. Such as: (Pia Gjellerup (udpeget af S))
#To avoid this, we first find all that have "udpeget af" in them, and then remove everything after the first "(" along with all parantheses. This is done by the function "remove_udpeget_af"

def remove_udpeget_af(df):
    #find all that have "udpeget af" or "Udpeget af" in them
    udpeget_af = df[df["navn"].str.contains("udpeget af|Udpeget af")]
    #remove everything after the first "(" along with all parantheses
    udpeget_af["navn"] = udpeget_af["navn"].str.split("(").str[0]
    #remove all parantheses
    udpeget_af["navn"] = udpeget_af["navn"].str.replace("(", "")
    udpeget_af["navn"] = udpeget_af["navn"].str.replace(")", "")

    #update the df with the new values
    df.update(udpeget_af)
    return df

df = remove_udpeget_af(df)
```

```python
from fuzzywuzzy import fuzz, process
from collections import defaultdict

def group_similar_entities(entity_list, similarity_threshold=85):
    groups = defaultdict(list)
    ungrouped_entities = set(entity_list)


    for entity in entity_list:
        if entity in ungrouped_entities:
            similar_entities = process.extract(entity, ungrouped_entities, limit=None, scorer=fuzz.token_set_ratio)
            similar_entities = [e[0] for e in similar_entities if e[1] >= similarity_threshold]
            groups[entity].extend(similar_entities)
            ungrouped_entities.difference_update(similar_entities)

    return dict(groups)
```

```python
def get_aliases_from_wikidata(entity_name, language="da"):
    sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
    query = f"""
    SELECT ?alias WHERE {{
        ?entity rdfs:label "{entity_name}"@da.
        ?entity skos:altLabel ?alias.
        FILTER (lang(?alias) = "da")
    }}
    """
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

    aliases = [result["alias"]["value"] for result in results["results"]["bindings"]]
    return {entity_name: aliases}
```

```python
import time
from tqdm import tqdm
from retry import retry
from SPARQLWrapper import SPARQLWrapper, JSON

########


# def get_aliases_with_retry(entity_name, language="da", retries=3):
#     for i in range(retries):
#         try:
#             aliases = get_aliases_from_wikidata(entity_name, language)
#             return aliases
#         except Exception as e:
#             if i < retries - 1:
#                 time.sleep(2 ** i)
#             else:
#                 raise e

@retry(tries=5, delay=2, backoff=2)
def get_aliases_with_retry(entity_name, language="da"):
    return get_aliases_from_wikidata(entity_name, language)

def create_entity_alias_map(df, institution_id_map, grouped_institutions):
    # Initialize the entity_alias_map with all entities from the dataframe
    entity_alias_map = {entity_id: {"same_named_ids": set(), "aliases": set()} for entity_id in df['id'].unique()}

    # Process "Person" entities
    person_df = df[df['entity_type'] == 'person']
    for _, row in tqdm(person_df.iterrows(), desc="Processing Persons", total=len(person_df)):
        name, entity_id = row['navn'], row['id']
        same_name_entities = person_df[person_df['navn'] == name]['id'].tolist()
        entity_alias_map[entity_id]["same_named_ids"].update(same_name_entities)

        # Add Wikidata aliases
        name = name.replace('"', "'")  # Replace double quotes with single quotes
        wikidata_aliases = get_aliases_with_retry(name, language="da")
        for aliases in wikidata_aliases.values():
            entity_alias_map[entity_id]["aliases"].update(aliases)

    # Process "Institution" entities
    for main_name, similar_names in tqdm(grouped_institutions.items(), desc="Processing Institutions", total=len(grouped_institutions)):
        main_entity_id = institution_id_map[main_name]
        entity_alias_map[main_entity_id]["same_named_ids"].update(institution_id_map[name] for name in similar_names)

        # Add Wikidata aliases
        for name in similar_names:
            name = name.replace('"', "'")  # Replace double quotes with single quotes
            wikidata_aliases = get_aliases_with_retry(name, language="da")
            for aliases in wikidata_aliases.values():
                entity_alias_map[main_entity_id]["aliases"].update(aliases)

    return entity_alias_map


```
